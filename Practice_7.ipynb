Це називається навчанням без нагляду.

**Кластеризація K-середніх** є найпопулярнішим алгоритмом навчання без нагляду. Він використовується, коли у нас є дані без міток, які є даними без визначених категорій або груп. Алгоритм дотримується легкого або простого способу класифікації даного набору даних через певну кількість кластерів, фіксованих попередньо. Алгоритм K-Means працює ітеративно, щоб призначити кожну точку даних одній із K груп на основі наданих функцій. Точки даних групуються на основі подібності ознак.

## Кластеризація інтуїції k-середніх

Кластеризація K-середніх використовується для пошуку внутрішніх груп у немаркованому наборі даних і отримання висновків з них. Він заснований на кластеризації на основі центроїдів.

Центроїд - центроїд - це точка даних у центрі кластера. У кластеризації на основі центроїдів кластери представлені центроїдом. Це ітераційний алгоритм, у якому поняття подібності випливає з того, наскільки близька точка даних до центроїда кластера. Кластеризація K-Means працює наступним чином: Алгоритм кластеризації K-Means використовує ітераційну процедуру для отримання кінцевого результату. Алгоритм вимагає кількості кластерів K і набору даних як вхідних даних. Набір даних — це набір функцій для кожної точки даних. Алгоритм починається з початкових оцінок для K центроїдів. Потім алгоритм виконує ітерації між двома кроками:

## Крок призначення даних

Кожен центроїд визначає один із кластерів. На цьому кроці кожній точці даних приписується її найближчий центр, який базується на квадраті евклідової відстані. Отже, якщо ci — це сукупність центроїдів у наборі C, то кожна точка даних призначається кластеру на основі мінімальної евклідової відстані.

## Крок оновлення центроїдів

На цьому кроці центроїди перераховуються й оновлюються. Для цього береться середнє значення всіх точок даних, призначених кластеру цього центроїда.

Потім алгоритм повторює кроки 1 і 2, поки не буде виконано критерій зупинки. Критерії зупинки означають, що точки даних не змінюють кластери, сума відстаней мінімізується або досягається деяка максимальна кількість ітерацій. Цей алгоритм гарантовано сходиться до результату. Результатом може бути локальний оптимум, що означає, що оцінка більш ніж одного запуску алгоритму з рандомізованими початковими центроїдами може дати кращий результат.

## Вибір значення К

Алгоритм K-середніх залежить від пошуку кількості кластерів і міток даних для попередньо визначеного значення K. Щоб знайти кількість кластерів у даних, нам потрібно запустити алгоритм кластеризації K-Means для різних значень K і порівняти результати. Отже, продуктивність алгоритму K-Means залежить від значення K. Ми повинні вибрати оптимальне значення K, яке дає нам найкращу продуктивність. Існують різні методики для визначення оптимального значення K. Найпоширенішим методом є метод ліктя, який описано нижче.

## Метод ліктя

Для визначення оптимальної кількості кластерів у кластеризації K-середніх використовується метод коліна. Метод коліно відображає значення функції вартості, отримане різними значеннями K. На схемі нижче показано, як працює метод ліктя:

![Elbow method in K-Means](https://www.oreilly.com/library/view/statistics-for-machine/9781788295758/assets/995b8b58-06f1-4884-a2a1-f3648428e947.png)

Ми бачимо, що якщо K збільшується, середнє спотворення зменшиться. Тоді кожен кластер матиме менше складових екземплярів, і екземпляри будуть ближче до своїх відповідних центроїдів. Однак покращення середнього спотворення зменшаться зі збільшенням K. Значення K, при якому поліпшення спотворення зменшується найбільше, називається колінцем, при якому ми повинні припинити поділ даних на наступні кластери.

### Інерційність

* Інерція не є нормованим показником.

* Нижчі значення інерції кращі, а нуль – оптимальний.

* Але в просторах з дуже високою вимірністю евклідові відстані мають тенденцію збільшуватися (це приклад прокляття розмірності).

## Використаємо метод ліктя, щоб знайти оптимальну кількість кластерів

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

* За наведеним вище графіком ми бачимо, що існує перелом при k=2.

* Отже, k=2 можна вважати гарною кількістю кластера для кластерування цих даних.

* Але ми бачили, що я досяг слабкої точності класифікації 1% з k=2.

* Я знову напишу необхідний код з k=2 для зручності.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2,random_state=0)

kmeans.fit(X)

labels = kmeans.labels_

# check how many of the samples were correctly labeled

correct_labels = sum(y == labels)

print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))

print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

Отже, наша слабка неконтрольована модель класифікації досягла дуже слабкої точності класифікації в 1%.

Я перевірю точність моделі з різною кількістю кластерів.

### Початкова розвідка

У цьому розділі ми об’єднаємо дані в більш корисні показники. Зокрема, ми обчислимо загальну кількість поїздок на годину для кожної години місяця, а потім використаємо ці дані для обчислення середньої кількості поїздок за кожну годину в певний день тижня.

#### Дані погодинної поїздки

Ми будемо рахувати поїздки за кожну годину місяця та збережемо функцію буднього дня. Операція `groupby` призначає ієрархічний індекс, який ми можемо скинути. Крім того, ми перейменуємо стовпець, щоб вказати, що це підрахунок загальної кількості поїздок, які відповідають критеріям.

```python
## groupby operation
hourly_ride_data = rides.groupby(['day','hour','weekday'])['timestamp'].count()

## reset index
hourly_ride_data = hourly_ride_data.reset_index()

## rename column
hourly_ride_data = hourly_ride_data.rename(columns = {'timestamp':'ride_count'})

## ocular analysis
hourly_ride_data.head()
```

#### Середні погодинні показники дня тижня

Далі ми будемо використовувати погодинні дані для обчислення середньогодинного значення для кожного дня тижня. Ми відсортуємо дані за попередньо визначеним індексом днів тижня для чистого графіка.

```python
## groupby operation
weekday_hourly_avg = hourly_ride_data.groupby(['weekday','hour'])['ride_count'].mean()

## reset index
weekday_hourly_avg = weekday_hourly_avg.reset_index()

## rename column
weekday_hourly_avg = weekday_hourly_avg.rename(columns = {'ride_count':'average_rides'})

## sort by categorical index
weekday_hourly_avg = weekday_hourly_avg.sort_index()

## ocular analysis
weekday_hourly_avg.head()
```

### Просторова візуалізація

Чи можемо ми щось сказати про розташування цих поїздок?

Давайте створимо функцію, яка допоможе нам нанести дані на карту. Ми будемо використовувати пакет Basemap для збирання результатів на основі місцезнаходження для створення теплової карти. Basemap використовуватиме метод hexbin з matplotlib для просторового об’єднання. Ми встановимо параметр `zorder` так, щоб теплова карта була нанесена поверх заливки континенту.

```python
def heat_map(ax_loc,title_str,rides_this_hour,nsew):
    
    ## get the axis
    ax = fig.add_subplot(ax_loc)

    ## make the basemap object
    m = Basemap(projection='merc', urcrnrlat=nsew[0], llcrnrlat=nsew[1],
                urcrnrlon=nsew[2], llcrnrlon=nsew[3], lat_ts=nsew[1], resolution='f')

    ## draw the background features
    m.drawmapboundary(fill_color = 'xkcd:light blue')
    m.fillcontinents(color='xkcd:grey', zorder = 1)
    m.drawcoastlines()
    m.drawrivers()

    ## project the GPS coordinates onto the x,y representation
    x, y = m(rides_this_hour['lon'].values, rides_this_hour['lat'].values)

    ## count the instances using the hexbin method and plot the results
    m.hexbin(x, y, gridsize=1000, mincnt = 1, bins = 'log', cmap=cm.YlOrRd, zorder = 2);

    ## set the title
    ax.set_title(title_str, fontsize=24)
```

```python
## set weekday for analysis
target_day = 'Thursday'

## north,south,east,west lat/lon coordinates for bounding box
nsew = [40.9, 40.6, -73.8, -74.1]

## create figure
fig = plt.figure(figsize=(14,8))

## target hours
hrs = [17, 21, 23]

## axis subplot locations
ax_loc = [131, 132, 133] 

## title strings
title_str = ['5 PM', '9 PM', '11 PM']

## plot loop
for ii in range(len(ax_loc)):

    ## get the ride data from the target hour
    rides_this_hour = rides.loc[(rides['weekday'] == target_day) & (rides['hour'] == hrs[ii])]

    ##  plot the heat map
    heat_map(ax_loc[ii],title_str[ii],rides_this_hour,nsew)
```

Функція кластеризації
Спочатку ми напишемо функцію, яка запускає алгоритм кластеризації та повертає "гарячі точки". Ми отримаємо координати центроїда та кількість підхоплювачів у кожному кластері.

```python
def get_hot_spots(max_distance,min_cars,ride_data):
    
    ## get coordinates from ride data
    coords = ride_data.as_matrix(columns=['lat', 'lon'])
    
    ## calculate epsilon parameter using
    ## the user defined distance
    kms_per_radian = 6371.0088
    epsilon = max_distance / kms_per_radian
    
    ## perform clustering
    db = DBSCAN(eps=epsilon, min_samples=min_cars,
                algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
    
    ## group the clusters
    cluster_labels = db.labels_
    num_clusters = len(set(cluster_labels))
    clusters = pd.Series([coords[cluster_labels == n] for n in range(num_clusters)])
    
    ## report
    print('Number of clusters: {}'.format(num_clusters))
    
    ## initialize lists for hot spots
    lat = []
    lon = []
    num_members = []
    
    ## loop through clusters and get centroids, number of members
    for ii in range(len(clusters)):

        ## filter empty clusters
        if clusters[ii].any():

            ## get centroid and magnitude of cluster
            lat.append(MultiPoint(clusters[ii]).centroid.x)
            lon.append(MultiPoint(clusters[ii]).centroid.y)
            num_members.append(len(clusters[ii]))
            
    hot_spots = [lon,lat,num_members]
    
    return hot_spots
```

```python
## Function for various clusters
def compute_cluster (clType,df= X):
    if clType=='KMeans':
        result = cluster.KMeans(n_clusters= 2).fit_predict(df) 
    elif clType == 'MeanShift' :
        result = cluster.MeanShift(bandwidth=0.2).fit_predict(df)    
    elif clType == 'MiniBatchKMeans':
        result = cluster.MiniBatchKMeans(n_clusters=2).fit_predict(df)        
    elif clType == 'SpectralClustering':
        result = cluster.SpectralClustering(n_clusters=2).fit_predict(df)
    elif clType == 'DBSCAN':
        result = cluster.DBSCAN(eps=0.3).fit_predict(df)
    elif clType == 'Affinity Propagation':
        result = cluster.AffinityPropagation(damping=0.9, preference=-200).fit_predict(df)
    elif clType == 'Birch':
        result = cluster.Birch(n_clusters= 2).fit_predict(df)
    elif clType == 'GaussianMixture' :
        gmm = mixture.GaussianMixture( n_components=2, covariance_type='full')
        gmm.fit(df)
        result = gmm.predict(df)  
    else:
        print("exit")
    
    cl_df.loc[cl_df.Name == clType, 'Silhouette-Coeff'] = metrics.silhouette_score(df, result, metric='euclidean')
    cl_df.loc[cl_df.Name == clType, 'Calinski-Harabaz'] = metrics.calinski_harabasz_score(df, result)
    
    return result
```

**Коефіцієнт силуету** (sklearn.metrics.silhouette_score) є прикладом такого оцінювання, де вищий показник коефіцієнта силуету відноситься до моделі з краще визначеними кластерами. Коефіцієнт Силуету визначається для кожного зразка і складається з двох балів:

a: Середня відстань між зразком і всіма іншими точками в тому ж класі. b: середня відстань між вибіркою та всіма іншими точками в наступному найближчому кластері. Тоді коефіцієнт силуету s для окремого зразка задається так: s = {b - a}/{max(a, b)}

Оцінка обмежена між -1 для неправильної кластеризації та +1 для дуже щільної кластеризації. Оцінки біля нуля вказують на перекриття кластерів. Оцінка вища, коли кластери щільні та добре розділені, що відноситься до стандартної концепції кластера.

індекс **Калінського-Харабаза** (sklearn.metrics.calinski_harabaz_score) можна використовувати для оцінки моделі, де вищий бал Калінського-Харабаза відноситься до моделі з краще визначеними кластерами.

Для k кластерів оцінка Калінського-Харабаза s наводиться як відношення середньої дисперсії між кластерами та дисперсії всередині кластера. Оцінка вища, коли кластери щільні та добре розділені, що відноситься до стандартної концепції кластера.
